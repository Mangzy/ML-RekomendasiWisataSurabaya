```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import bnlearn as bn

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from pgmpy.models import BayesianNetwork

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('kbmg.csv')

train['Passenger ID'] = train.index

# Atau, tambahkan kolom "Passenger ID" dengan nilai unik (misalnya, nomor urut)
train['Passenger ID'] = range(1, len(train) + 1)

cols = list(train.columns)
cols = ['Passenger ID'] + [col for col in cols if col != 'Passenger ID']
train = train[cols]

train.set_index('Passenger ID', inplace=True)

# Tampilkan trainFrame
train
```

```python
drop_list = ['Aktivitas Wisata', 'Nama Wisata']

train_wisata = train.drop(columns=drop_list)
train_wisata
```

```python
dfhot_train, dfnum_train = bn.df2onehot(train_wisata)
```

```python
dfnum_target = dfnum_train.pop('Kebun Raya Mangrove Gunung Anyar')
dfnum_target
```

```python
Xtrain, Xval, Ztrain, Zval = train_test_split(dfnum_train, dfnum_target, test_size=0.2, random_state=0)
valid = pd.concat([Xval, Zval], axis='columns')
dfnum = pd.concat([Xtrain, Ztrain], axis='columns')
dfnum
```

```python
edges = [
    ('Escape', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Relaxation', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Play', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Strenghthening family bonds', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Prestige', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Social Interaction', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Romance', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Educational Opportunity', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Self-fulfilment', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Wish-fulfiment', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Lingkungan', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Infrastruktur', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Fasilitas', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Akomodasi', 'Kebun Raya Mangrove Gunung Anyar'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Makan Bersama'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Berolahraga'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Belajar'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Berinteraksi dengan satwa'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Mengambil Foto'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Beribadah'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Berkemah'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Melihat Pemandangan'),
    ('Kebun Raya Mangrove Gunung Anyar', 'Berbelanja')
]
```

```python
params_static = {
    'figsize': (15, 10),
    'width': 20,  # Menyesuaikan lebar panah
    'edge_alpha': 0.5,  # Menyesuaikan kejelasan panah
    'arrowstyle': '-|>',  # Gaya panah (ubah sesuai kebutuhan)
    'arrowsize': 20,  # Ukuran panah
    'minscale': 5,
    'maxscale': 15,
    'layout': 'spring_layout',
    'font_size': 14,
    'font_family': 'sans-serif',
    'alpha': 0.8,
    'node_shape': 'o',
    'font_color': '#000000',
    'facecolor': 'white',
    'visible': True
}
```

```python
DAG = bn.make_DAG(edges, verbose=0)


bn.print_CPD(DAG)


# Plot
G = bn.plot(DAG, interactive=False, params_static=params_static)


# Parameter learning
model = bn.parameter_learning.fit(DAG, dfnum, verbose=3);

```

```python
def get_acc(model, df, col):
    # Get accuracy score by the model for the validation dataset df with target col
    pred = bn.predict(model, df, variables=[col])
    acc = accuracy_score(df[col], pred[col])
    print('Accuracy -', acc)
    return acc
```

```python
get_acc(model, valid, 'Kebun Raya Mangrove Gunung Anyar')
```

```python
pred = bn.predict(model, valid, variables=['Kebun Raya Mangrove Gunung Anyar'])
```

```python
# create confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ""})
ConfusionMatrixDisplay(confusion_matrix(valid['Kebun Raya Mangrove Gunung Anyar'], pred['Kebun Raya Mangrove Gunung Anyar']), display_labels=valid['Kebun Raya Mangrove Gunung Anyar'].unique()).plot()
```
