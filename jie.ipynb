```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import bnlearn as bn

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from pgmpy.models import BayesianNetwork

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('jie.csv')

train['Passenger ID'] = train.index

# Atau, tambahkan kolom "Passenger ID" dengan nilai unik (misalnya, nomor urut)
train['Passenger ID'] = range(1, len(train) + 1)

cols = list(train.columns)
cols = ['Passenger ID'] + [col for col in cols if col != 'Passenger ID']
train = train[cols]

train.set_index('Passenger ID', inplace=True)

# Tampilkan trainFrame
train
```

```python
drop_list = ['Aktivitas Wisata', 'Nama Wisata']

train_wisata = train.drop(columns=drop_list)
train_wisata
```

```python
dfhot_train, dfnum_train = bn.df2onehot(train_wisata)
```

```python
dfnum_target = dfnum_train.pop('Jatim International Expo (JIE) Convention Exhibiton')
dfnum_target
```

```python
Xtrain, Xval, Ztrain, Zval = train_test_split(dfnum_train, dfnum_target, test_size=0.2, random_state=0)
valid = pd.concat([Xval, Zval], axis='columns')
dfnum = pd.concat([Xtrain, Ztrain], axis='columns')
dfnum
```

```python
edges = [
    ('Escape', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Relaxation', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Play', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Strenghthening family bonds', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Prestige', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Social Interaction', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Romance', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Educational Opportunity', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Self-fulfilment', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Wish-fulfiment', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Lingkungan', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Infrastruktur', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Fasilitas', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Akomodasi', 'Jatim International Expo (JIE) Convention Exhibiton'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Makan Bersama'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Berolahraga'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Belajar'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Berinteraksi dengan satwa'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Mengambil Foto'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Beribadah'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Berkemah'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Melihat Pemandangan'),
    ('Jatim International Expo (JIE) Convention Exhibiton', 'Berbelanja')
]
```

```python
params_static = {
    'figsize': (15, 10),
    'width': 20,  # Menyesuaikan lebar panah
    'edge_alpha': 0.5,  # Menyesuaikan kejelasan panah
    'arrowstyle': '-|>',  # Gaya panah (ubah sesuai kebutuhan)
    'arrowsize': 20,  # Ukuran panah
    'minscale': 5,
    'maxscale': 15,
    'layout': 'spring_layout',
    'font_size': 14,
    'font_family': 'sans-serif',
    'alpha': 0.8,
    'node_shape': 'o',
    'font_color': '#000000',
    'facecolor': 'white',
    'visible': True
}
```

```python
DAG = bn.make_DAG(edges, verbose=0)


bn.print_CPD(DAG)


# Plot
G = bn.plot(DAG, interactive=False, params_static=params_static)


# Parameter learning
model = bn.parameter_learning.fit(DAG, dfnum, verbose=3);

```

```python
def get_acc(model, df, col):
    # Get accuracy score by the model for the validation dataset df with target col
    pred = bn.predict(model, df, variables=[col])
    acc = accuracy_score(df[col], pred[col])
    print('Accuracy -', acc)
    return acc
```

```python
get_acc(model, valid, 'Jatim International Expo (JIE) Convention Exhibiton')
```

```python
pred = bn.predict(model, valid, variables=['Jatim International Expo (JIE) Convention Exhibiton'])
```
